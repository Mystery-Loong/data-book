# 10.1 GroupBy framework

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"key1": ["a", "a", None, "b", "b", "a", None],
                   "key2": pd.Series([1, 2, 1, 2, 1, None, 1], dtype="Int64"),
                   "data1": np.random.standard_normal(7),
                   "data2": np.random.standard_normal(7)})
```

```python
df
```

```python
grouped = df["data1"].groupby(df["key1"])
```

```python
grouped
```

```python
grouped.mean()
```

```python
means = df["data1"].groupby([df["key1"], df["key2"]]).mean()
```

```python
means
```

```python
means.unstack()
```

```python
states = np.array(["OH", "CA", "CA", "OH", "OH", "CA", "OH"])
```

```python
years = [2005, 2005, 2006, 2005, 2006, 2005, 2006]
```

```python
df["data1"].groupby([states, years]).mean()
```

```python
df.groupby("key1").mean()
```

```python
df.groupby(["key1", "key2"]).mean()
```

```python
df.groupby(["key1", "key2"]).size()
```

```python
df.groupby("key1", dropna=False).size()
```

```python
df.groupby(["key1", "key2"], dropna=False).size()
```

```python
df.groupby("key1").count()
```

```python
df.groupby("key2").count()
```

# 10.1.1 iterate on grouping

```python
for name, group in df.groupby("key1"):
    print(name)
    print(group)
```

```python
for (k1, k2), group in df.groupby(["key1", "key2"]):
    print((k1, k2))
    print(group)
```

```python
pieces = {name: group for name, group in df.groupby("key1")}
```

```python
pieces["b"]
```

```python
grouped = df.groupby({"key1": "key", "key2": "key",
                      "data1": "data", "data2": "data"}, axis="columns")
```

```python
for group_key, group_values in grouped:
    print(group_key)
    print(group_values)
```

# 10.1.3 grouping using dictionaries and series 

```python
people = pd.DataFrame(np.random.standard_normal((5, 5)),
                      columns=["a", "b", "c", "d", "e"],
                      index=["Joe", "Steve", "Wanda", "Jill", "Trey"])
```

```python
people.iloc[2:3, [1, 2]] = np.nan #join some none
```

```python
people
```

```python
mapping = {"a": "red", "b": "red", "c": "blue",
           "d": "blue", "e": "red", "f": "orange"}
```

```python
by_column = people.groupby(mapping, axis="columns")
```

```python
by_column.sum()
```

```python
map_series = pd.Series(mapping)
```

```python
map_series
```

```python
people.groupby(map_series, axis="columns").count()
```

# 10.1.5 grouping according to index level

```python
columns = pd.MultiIndex.from_arrays([["US", "US", "US", "JP", "JP"],
                                     [1, 3, 5, 1, 3]],
                                    names=["cty", "tenor"])
```

```python
hier_df = pd.DataFrame(np.random.standard_normal((4, 5)), columns=columns)
```

```python
hier_df
```

```python
hier_df.groupby(level="cty", axis="columns").count()
```

# 10.2.1 column by column operations and multi function applications

```python
tips = pd.read_csv("examples/tips.csv")
```

```python
tips.head()
```

```python
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
```

```python
tips.head()
```

```python
grouped = tips.groupby(["day", "smoker"])
```

```python
grouped_pct = grouped["tip_pct"]
```

```python
grouped_pct.agg("mean")
```

```python
grouped_pct.agg(["mean", "std", peak_to_peak])
```

```python
grouped_pct.agg([("average", "mean"), ("stdev", np.std)])
```

```python
functions = ["count", "mean", "max"]
```

```python
result = grouped[["tip_pct", "total_bill"]].agg(functions)
```

```python
result
```

```python
result["tip_pct"]
```

```python
ftuples = [("Average", "mean"), ("Variance", np.var)]
```

```python
grouped[["tip_pct", "total_bill"]].agg(ftuples)
```

```python
grouped.agg({"tip": np.max, "size": "sum"})
```

```python
grouped.agg({"tip_pct": ["min", "max", "mean", "std"],
             "size": "sum"})
```

# 10.3 Apply: universal "split application union" specification

```python
def top(df, n=5, column="tip_pct"):
    return df.sort_values(column, ascending=False)[:n]
```

```python
top(tips, n=6)
```

```python
tips.groupby("smoker").apply(top)
```

```python
tips.groupby(["smoker", "day"]).apply(top, n=1, column="total_bill")
```

```python
result = tips.groupby("smoker")["tip_pct"].describe()
```

```python
result
```

```python
result.unstack("smoker")
```

```python
def f(group):
    return group.describe()
```

```python
grouped.apply(f)
```
